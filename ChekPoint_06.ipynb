# KNN

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
df=load_iris()
X=df.data
y=df.target
```

# graphique de la dataset IRIS

```python
plt.figure(figsize=(12,9))
plt.scatter(X[:,0],X[:,1],c=y,alpha=0.5,s=X[:,2]*100)
plt.show()
```

# Split Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test=train_test_split(X,y, test_size=.25, random_state=10)
```

```python
from math import sqrt
sqrt(len(X_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,f1_score,accuracy_score,precision_score,recall_score
```

```python
m=[]
for i in range(1,10):
    knclf=KNeighborsClassifier(n_neighbors=i)
    knclf.fit(X_train,y_train)
    m.append( knclf.score(X_test,y_test))
plt.figure(figsize=(12,8))
plt.plot(m)
plt.xlabel("K")
plt.ylabel("score")
plt.show()
```

```python
knclf=KNeighborsClassifier(n_neighbors=6)
knclf.fit(X_train,y_train)
y_pred=knclf.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
plt.figure(figsize=(12,8))
sns.heatmap(cm, annot=True, fmt=".1f")
plt.title("Matircie de confusion")
plt.show()
```

```python
scr=knclf.score(X_test,y_test)*100
accscr=accuracy_score(y_test,y_pred)*100
f1scr=f1_score(y_test,y_pred,average="macro")*100
pescr=precision_score(y_test,y_pred,average="macro")*100
rescr=recall_score(y_test,y_pred,average="macro")*100
```

```python
print("Les metrics: \n"
     "\t score=",scr,"%\n",
     "\t accuracy_score=",accscr,"%\n",
     "\t f1_score=",f1scr,"%\n",
     "\t precision_score=",pescr,"%\n",
     "\t recall_score=",rescr,"%\n",)
```
